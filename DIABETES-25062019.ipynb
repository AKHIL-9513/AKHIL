```python
import pandas as pd
```

```python
df=pd.read_csv("diabetes.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
# univariate plot eg-histogram
# bivariate plot eg-scatter
df.boxplot(by='Age', column=["BloodPressure"], grid =True, figsize=(20,10))
```

```python
# To see how Pregnancies & Glucose are related:
# Making bivariate plot:
x = df["Pregnancies"]
y = df["Glucose"]
```

```python
df.plot.bar(x=["Pregnancies"],y=["Glucose"])
```

```python
pregnancies=df["Pregnancies"]
```

```python
plt.hist(pregnancies,bins=50,align='mid',histtype='bar', density=True,color="green")
plt.ylabel("BMI")
plt.xlabel("Pregnancies")
plt.title("Histogram of Pregnancies")
plt.grid(True)
plt.show()
```
